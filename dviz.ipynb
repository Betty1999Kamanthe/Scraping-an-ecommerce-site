### Import necessary libraries

```python
from bs4 import BeautifulSoup
import requests
from csv import writer
from urllib.request import Request, urlopen
import urllib.request
from urllib.parse import urlparse 
```

### User can input choose between Dealer and Private as mentioned in the screenshot .

```python
choice = input("Choose one please: \n1. Dealer \n2. Private")

print("Your choice is: "+choice)
```

```python
if choice== 1:
    url = 'https://www.bilbasen.dk/brugt/bil?PriceFrom=0&ZipCode=0&IncludeEngrosCVR=True&Seller=1&IncludeSellForCustomer=True&IncludeWithoutVehicleRegistrationTax=True&IncludeLeasing=True&HpFrom=&HpTo='
else:
    url = 'https://www.bilbasen.dk/brugt/bil?PriceFrom=0&ZipCode=0&IncludeEngrosCVR=True&Seller=2&IncludeSellForCustomer=True&IncludeWithoutVehicleRegistrationTax=True&IncludeLeasing=True&HpFrom=&HpTo='
    
```

```python
#we disguise our request by sending a HTTP Header indicating the request was being sent from the Chrome web browser.
req = Request(url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'})

page = urlopen(req).read()

#print(page)
```

```python
#run the html document through beautifulsoup in order to get the Beautifulsoup object
soup = BeautifulSoup(page , 'html.parser')
```

### First stage you scrape all the listings and the URLs

```python
#inspect element in your web browser as well to see where the specific part for the web page you want to scrape falls within in the HTML document.
cars = soup.find_all('div' , class_ = 'row listing listing-plus bb-listing-clickable')
```

```python
#store hrefs into a dict so that we can loop over later
ind_href = []

with open ('dealer.csv','w', encoding = 'utf8',newline = '') as f:
    thewriter = writer(f)
    header = ['Name','Href']
    thewriter.writerow(header)
    
    #Loop over to get the urls and cars
    for car in cars:
        Name = car.find('a', class_="listing-heading darkLink").text.replace('\n', '')
        href = car.a['href']
        if not urlparse(href).scheme: #Makes sure the url is in the correct format
            href= 'https://www.bilbasen.dk' + href

        ind_href.append(href)#append the href to ind_href

        info = [Name,href]

        thewriter.writerow(info)

        print("Info..")
        print(info)

#print(ind_href)
    
```

## Stage 2 scrape individual cars. 

###### Basic Information

```python
with open ('dealerinfo.csv','w', encoding = 'utf8',newline = '') as a:
    thewriter = writer(a)
    header = ['Name','Elements','Specification','DealerContact','PrivateContact']
    thewriter.writerow(header)

    for x in ind_href:

        #we disguise our request by sending a HTTP Header indicating the request was being sent from the Chrome web browser.
        req2 = Request(x, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36'})

        page2 = urlopen(req2).read()
        #print(page2)

        #run the html document through beautifulsoup in order to get the Beautifulsoup object
        soup2 = BeautifulSoup(page2 , 'html.parser')

        #inspect element in your web browser as well to see where the specific part for the web page you want to scrape falls within in the HTML document.
        ind_cars = soup2.find_all('div' , class_ = 'bbVipWrapper cf')

        #print (x)


        #Loop to find the name and basic information about each car
        for ind_car in ind_cars:
            Name2 = ind_car.h1.get_text().replace('\n',"").replace('\r', '')

            content =ind_car.find_all('span', class_="value")
            elements = [tag.text.strip() for tag in content]
            
            #Detailed Information
            import pandas as pd
            # Defining of the dataframe
            df = pd.DataFrame(columns=['Specifications', 'Showing_car', 'Your_car'])

            tables = ind_car.find('table', class_='type1')

            table = tables.find_all('tr')

            detailed = []
            #Loop over the specifications table
            
            for row in table:    
                # Find all data for each column
                columns = row.find_all('td')

                if(columns != []):
                    Specifications = columns[0].text.strip()
                    Showing_car = columns[-2].text.strip()
                    Your_car = columns[-1]

                    df = df.append({'Specifications': Specifications,  'Showing_car': Showing_car, 'Your_car': Your_car}, ignore_index=True)
                    detailed = [df]
            
            
            #Contact Information
            # Dealer Contact
            dcontact = ind_car.find_all('span', class_="phone-numbers")
            dealer_contact =[y.a['href']for y in dcontact]
            
            #private contact
            pcontact = ind_car.find_all('div', class_="phoneNumbers")
            private_contact =[x.a['href']for x in pcontact]

            info2 = [Name2,elements,detailed,dealer_contact,private_contact]


            thewriter.writerow(info2)

            print("Info.. ")
            print(info2)
```
